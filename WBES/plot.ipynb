```python
import matplotlib.pyplot as plt
import numpy as np

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d} \n ({p:.2f}%)'.format(p=pct,v=val)
    return my_autopct
```

```python
values = np.array([14284, 162868])
# 设置画布大小
plt.figure(figsize=(6,3))
figure,axes = plt.subplots(1,1,figsize = (6,6),dpi = 80) 
plt.pie(values,
        labels=['Surveys with \n outcome','Surveys without \n outcome'], # 设置饼图标签
        colors=["#d5695d", "#5d8ca8"], # 设置饼图颜色
        explode=(0, 0.02), # 第二部分突出显示，值越大，距离中心越远
        textprops={'fontsize': 14}, # 设置文本标签大小和颜色
        autopct=make_autopct(values), # 格式化输出百分比
       )
plt.title("Overall of Surveys", fontsize = 18) # 设置标题
plt.savefig('WBES.png') # 保存图片
plt.show()
```
